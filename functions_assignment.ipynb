1. What is the difference between a function and a method in Python?
-> Function:

A function is independent and can be used anywhere in the program.

It is not tied to any object.

 Example of a function:

def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Calling the function
 Here, greet() is a function that works on its own.

 Method:

A method is a function inside a class that works only with objects.

It is called using dot notation (object.method()).

Example of a method:

class Person:
    def __init__(self, name):
        self.name = name  

    def greet(self):  # Method inside a class
        return f"Hello, {self.name}!"

p = Person("Alice")  
print(p.greet())  # Calling the method

7. What are the advantages of using generators over regular functions?
->Memory Efficient: No need to store all values in memory.

Lazy Evaluation: Values are generated one at a time.

Improved Performance: Faster execution in large data processing.

def count_up():  
    n = 1  
    while n <= 3:  
        yield n  
        n += 1  

counter = count_up()  
print(next(counter))  # Output: 1

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.

def sum_even_numbers(numbers):
    sum_even = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
    return int(sum_even)
sum_even_numbers([80])
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

def check_primes_in_range(limit):
    for num in range(1, limit + 1):
        if is_prime(num):
            print(f"{num} is prime")
        else:
            print(f"{num} is not prime")

check_primes_in_range(200)

# Print prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of term.
def fibonacci_iterator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a  # Yield the current Fibonacci number
        a, b = b, a + b  # Update values for the next iteration

# Using the iterator to generate the first 10 Fibonacci numbers
n_terms = 10  # Change this to any number of terms you want
fib_seq = fibonacci_iterator(n_terms)

# Print the Fibonacci sequence
print(list(fib_seq))


```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
power_of_2(5)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple
# List of tuples (e.g., (name, age))
data = [("Alice", 25), ("Bob", 20), ("Charlie", 30), ("David", 22)]

# Sorting based on the second element (age) using a lambda function
sorted_data = sorted(data, key=lambda x: x[1])

# Print the sorted list
print(sorted_data)

```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit
celsius_temperatures = [0, 10, 20, 30, 40]
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))
print(fahrenheit_temperatures)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, string))
remove_vowels("hello")
```
